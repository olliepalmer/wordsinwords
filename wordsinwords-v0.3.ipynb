```python
%%time

# wordsinwords v0.3

from tinydb import TinyDB, Query
db = TinyDB('db.json')
q = Query()
wordlist = 'words_alpha.txt'
# wordlist = 'words.txt'

# uncomment this if you want to start afresh
db.drop_tables()


def updateShort(short, long):
    if (db.get(q.word == short) == None):
        db.insert({'word':short,'in':[long],'contains':[]})
    else:
        in_ = db.get(q.word == short)['in']
        in_.append(long) if long not in in_ else in_
        db.upsert({'word':short,'in':in_},q.word==short)
        
def updateLong(short, long):
    if (db.get(q.word == long) == None):
        db.insert({'word':long,'in':[],'contains':[short]})
    else:
        contains_ = db.get(q.word == long)['contains']
        contains_.append(short) if short not in contains_ else contains_
        db.upsert({'word':long,'contains':contains_},q.word==long)

with open(wordlist,'r') as a:
    for short in a:
        short = short.rstrip("\r\n")
        with open(wordlist,'r') as b:
            for long in b:
                long = long.rstrip("\r\n")
                if (long.count(short) > 0 and long != short and len(short) > 1):
                    updateShort(short,long)
                    updateLong(short,long)
                    print (short,'is in',long)
```

```python
%%time

# marginally quicker way of doing things
# wordsinwords v0.3

from tinydb import TinyDB, Query
db = TinyDB('db2.json') # note started second db to test times etc
q = Query()
wordlist = 'words_alpha.txt'
# wordlist = 'words.txt'

# uncomment this if you want to start afresh
db.drop_tables()

with open(wordlist,'r') as a:
    for short in a:
        short = short.rstrip("\r\n")
        isin = []
        with open(wordlist,'r') as b:
            for long in b:
                long = long.rstrip("\r\n")
                if (long.count(short) > 0 and long != short and len(short) > 1):
                    isin.append(long)
        if len(isin) > 0:
            print(short,'>',len(isin))
            shorty = db.get(q.word==short)
            if(shorty==None):
                db.insert({'word':short,'in':isin,'contains':[]})
            else:
                in_ = shorty['in']
                for i in isin:
                    in_.append(i) if i not in in_ else in_
                db.upsert({'word':short,'in':isin},q.word==short)
            for i in isin:
                longy = db.get(q.word==i)
                if(longy==None):
                    db.insert({'word':i,'contains':[short],'in':[]})
                else:
                    contains_ = longy['contains']
                    contains_.append(short) if short not in contains_ else contains_
                    db.upsert({'word':i,'contains':contains_},q.word==i)
```
