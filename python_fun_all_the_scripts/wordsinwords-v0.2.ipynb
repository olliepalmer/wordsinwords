- 2 x words: _short_, _long_. for each _short_ in word_list, run through every other word in word_list that is longer to see if it can be found inside. if yes, append:
    - word: _short_ in: _long_
    - word: _long_ contains: _short_
- Database format: 'word' (word itself), 'in' (words this word is in), 'contains' (words this word contains)

```python
# now wrap it all together 2!

from tinydb import TinyDB, Query

db = TinyDB('wordsinwords.json')
User = Query()

blank = []
with open ('words_alpha.txt','r') as a:
    for short in a:
        with open ('words_alpha.txt','r') as b:
            for long in b:
                if (long.count(short) > 0 and long != short):
                    print (short,'is in',long)
                    if(db.count(User.word == short) < 1):
                        db.upsert({'word':short,'in':blank,'contains':blank},User.word == short)    
                    if(db.count(User.word == long) < 1):
                        db.upsert({'word':long,'in':blank,'contains':blank},User.word == long)
                    if (db.count(User.word == long) < 0):
                        contains = (db.get(User.word == long)['contains'])
                        contains.append(short) if short not in contains else contains
                        db.upsert({'word':long,'contains':contains},User.word='long')   
                    if (db.count(User.word == short) < 0):
                        in_ = (db.get(User.word == short)['in'])
                        in_.append(short) if short not in in_ else in_
                        db.upsert({'word':short,'in':in_},User.word='short')
```
